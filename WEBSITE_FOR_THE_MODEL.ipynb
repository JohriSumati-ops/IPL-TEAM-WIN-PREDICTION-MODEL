```python
import gradio as gr
import dill
import pandas as pd

# Load model
with open("pipe (1).dill", "rb") as f:
    pipe = dill.load(f)

teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]

cities = ['Hyderabad','Mumbai','Bangalore','Chennai','Kolkata','Delhi','Jaipur']

def predict_win(batting_team, bowling_team, city,
                runs_left, balls_left, wickets,
                target, crr, rrr):

    input_df = pd.DataFrame({
        'batting_team':[batting_team],
        'bowling_team':[bowling_team],
        'city':[city],
        'runs_left':[runs_left],
        'balls_left':[balls_left],
        'wickets':[wickets],
        'total_runs_x':[target],
        'crr':[crr],
        'rrr':[rrr]
    })

    result = pipe.predict_proba(input_df)

    return (
        f"Win Probability: {round(result[0][1]*100,2)}%",
        f"Loss Probability: {round(result[0][0]*100,2)}%"
    )
```

```python
def predict_win(batting_team, bowling_team, city,
                runs_left, balls_left, wickets,
                target, crr, rrr):

    input_df = pd.DataFrame({
        'batting_team':[batting_team],
        'bowling_team':[bowling_team],
        'city':[city],
        'runs_left':[runs_left],
        'balls_left':[balls_left],
        'wickets':[wickets],
        'total_runs_x':[target],
        'crr':[crr],
        'rrr':[rrr]
    })

    result = pipe.predict_proba(input_df)

    return (
        f"Win Probability: {round(result[0][1]*100,2)}%",
        f"Loss Probability: {round(result[0][0]*100,2)}%"
    )
```

```python
iface = gr.Interface(
    fn=predict_win,
    inputs=[
        gr.Dropdown(teams, label="Batting Team"),
        gr.Dropdown(teams, label="Bowling Team"),
        gr.Dropdown(cities, label="City"),
        gr.Number(label="Runs Left"),
        gr.Number(label="Balls Left"),
        gr.Number(label="Wickets Left"),
        gr.Number(label="Target Score"),
        gr.Number(label="Current Run Rate"),
        gr.Number(label="Required Run Rate")
    ],
    outputs=[
        gr.Textbox(label="Win Probability"),
        gr.Textbox(label="Loss Probability")
    ],
    title="üèè IPL Team Win Predictor",
    description="Predict IPL match outcome using Machine Learning"
)
```

```python
iface.launch(share=True)
```
